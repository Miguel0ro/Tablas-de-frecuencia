```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Generar 500 datos cualitativos aleatorios
objetos_posibles = ["mesa", "silla", "l√°mpara", "libro", "televisor", "computadora"]
datos_cualitativos = [random.choice(objetos_posibles) for _ in range(500)]  # 500 datos aleatorios

# Crear DataFrame
df_cualitativos = pd.DataFrame(datos_cualitativos, columns=['Objetos'])

# Tabla de frecuencia
tabla_frecuencia_cualitativos = df_cualitativos.value_counts().reset_index()
tabla_frecuencia_cualitativos.columns = ['Objeto', 'Frecuencia']

# Calcular frecuencia relativa
tabla_frecuencia_cualitativos['Frecuencia Relativa'] = tabla_frecuencia_cualitativos['Frecuencia'] / tabla_frecuencia_cualitativos['Frecuencia'].sum()

# Calcular frecuencia acumulada
tabla_frecuencia_cualitativos['Frecuencia Acumulada'] = tabla_frecuencia_cualitativos['Frecuencia'].cumsum()

# Mostrar la tabla de frecuencia
print("\nTabla de Frecuencia de Datos Cualitativos:")
print(tabla_frecuencia_cualitativos)

# Graficar frecuencia absoluta usando un histograma
df_cualitativos['Objetos'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Frecuencia Absoluta - Histograma')
plt.xlabel('Objetos')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Diagrama de barras horizontales
tabla_frecuencia_cualitativos.plot(kind='barh', x='Objeto', y='Frecuencia', color='orange')
plt.title('Frecuencia Absoluta - Diagrama de Barras')
plt.xlabel('Frecuencia')
plt.ylabel('Objetos')
plt.show()

# Graficar frecuencia relativa usando un pol√≠gono de frecuencia
plt.plot(tabla_frecuencia_cualitativos['Objeto'], tabla_frecuencia_cualitativos['Frecuencia Relativa'], marker='o')
plt.title('Frecuencia Relativa - Pol√≠gono de Frecuencia')
plt.xlabel('Objetos')
plt.ylabel('Frecuencia Relativa')
plt.xticks(rotation=45)
plt.show()

# Graficar frecuencia relativa usando un diagrama de pastel
plt.pie(tabla_frecuencia_cualitativos['Frecuencia Relativa'], labels=tabla_frecuencia_cualitativos['Objeto'], autopct='%1.1f%%', startangle=90)
plt.title('Frecuencia Relativa - Diagrama de Pastel')
plt.axis('equal')  # Para que sea un c√≠rculo
plt.show()

# Graficar frecuencia acumulada usando la ojiva
plt.plot(tabla_frecuencia_cualitativos['Objeto'], tabla_frecuencia_cualitativos['Frecuencia Acumulada'], marker='o', color='green')
plt.title('Frecuencia Acumulada - Ojiva')
plt.xlabel('Objetos')
plt.ylabel('Frecuencia Acumulada')
plt.xticks(rotation=45)
plt.show()
```
