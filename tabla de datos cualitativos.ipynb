```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Generar 500 datos cualitativos aleatorios
colores_posibles = ["rojo", "azul", "verde", "amarillo", "negro", "blanco"]
datos_cualitativos = [random.choice(colores_posibles) for _ in range(500)]  # 500 datos aleatorios

# Crear DataFrame
df_cualitativos = pd.DataFrame(datos_cualitativos, columns=['Colores'])

# Tabla de frecuencia
tabla_frecuencia_cualitativos = df_cualitativos.value_counts().reset_index()
tabla_frecuencia_cualitativos.columns = ['Color', 'Frecuencia']

# Calcular frecuencia relativa
tabla_frecuencia_cualitativos['Frecuencia Relativa'] = tabla_frecuencia_cualitativos['Frecuencia'] / tabla_frecuencia_cualitativos['Frecuencia'].sum()

# Calcular frecuencia acumulada
tabla_frecuencia_cualitativos['Frecuencia Acumulada'] = tabla_frecuencia_cualitativos['Frecuencia'].cumsum()

# Mostrar la tabla de frecuencia
print("\nTabla de Frecuencia de Datos Cualitativos:")
print(tabla_frecuencia_cualitativos)

# Graficar frecuencia absoluta usando un histograma
df_cualitativos['Colores'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Frecuencia Absoluta - Histograma')
plt.xlabel('Colores')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# Diagrama de barras horizontales
tabla_frecuencia_cualitativos.plot(kind='barh', x='Color', y='Frecuencia', color='orange')
plt.title('Frecuencia Absoluta - Diagrama de Barras')
plt.xlabel('Frecuencia')
plt.ylabel('Colores')
plt.show()

# Graficar frecuencia relativa usando un polígono de frecuencia
plt.plot(tabla_frecuencia_cualitativos['Color'], tabla_frecuencia_cualitativos['Frecuencia Relativa'], marker='o')
plt.title('Frecuencia Relativa - Polígono de Frecuencia')
plt.xlabel('Colores')
plt.ylabel('Frecuencia Relativa')
plt.xticks(rotation=45)
plt.show()

# Graficar frecuencia relativa usando un diagrama de pastel
plt.pie(tabla_frecuencia_cualitativos['Frecuencia Relativa'], labels=tabla_frecuencia_cualitativos['Color'], autopct='%1.1f%%', startangle=90)
plt.title('Frecuencia Relativa - Diagrama de Pastel')
plt.axis('equal')  # Para que sea un círculo
plt.show()

# Graficar frecuencia acumulada usando la ojiva
plt.plot(tabla_frecuencia_cualitativos['Color'], tabla_frecuencia_cualitativos['Frecuencia Acumulada'], marker='o', color='green')
plt.title('Frecuencia Acumulada - Ojiva')
plt.xlabel('Colores')
plt.ylabel('Frecuencia Acumulada')
plt.xticks(rotation=45)
plt.show()
```
