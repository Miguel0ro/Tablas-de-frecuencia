# Tabla de datos cuantitativos

```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar 500 datos aleatorios continuos entre 0 y 100
data = np.random.uniform(0, 100, 500)

# Crear un DataFrame para facilitar el manejo de datos
df = pd.DataFrame(data, columns=['Valores'])

# Mostrar la tabla de datos (sin mostrar todos los datos)
print("Tabla de datos generados:")
print(df.describe())

# Grfico de frecuencia absoluta (histograma)
plt.figure(figsize=(10, 6))
plt.hist(df['Valores'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()

# Grfico de barras horizontales
frecuencia = df['Valores'].value_counts(bins=10).sort_index()
plt.figure(figsize=(10, 6))
frecuencia.plot(kind='barh', color='orange')
plt.title('Diagrama de Barras Horizontales')
plt.xlabel('Frecuencia')
plt.ylabel('Valores')
plt.grid()
plt.show()

# Grfico de frecuencia relativa (polgono de frecuencia)
frecuencia_relativa = frecuencia / len(df)
plt.figure(figsize=(10, 6))
plt.plot(frecuencia_relativa.index.mid, frecuencia_relativa.values, marker='o', linestyle='-')
plt.title('Polgono de Frecuencia Relativa')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Relativa')
plt.grid()
plt.show()

# Diagrama de pastel
plt.figure(figsize=(10, 6))
plt.pie(frecuencia, labels=[f"{int(interval.left)}-{int(interval.right)}" for interval in frecuencia.index], autopct='%1.1f%%', startangle=90)
plt.title('Diagrama de Pastel')
plt.axis('equal')  # Para que el pastel sea circular
plt.show()

# Grfico de frecuencia acumulada (ojiva)
frecuencia_acumulada = frecuencia.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(frecuencia_acumulada.index.mid, frecuencia_acumulada.values, marker='o', linestyle='-')
plt.title('Ojiva - Frecuencia Acumulada')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulada')
plt.grid()
plt.show()
```
