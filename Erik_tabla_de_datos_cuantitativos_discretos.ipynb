```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Generar 500 datos aleatorios enteros entre 1 y 10
data = np.random.randint(1, 11, size=500)
df = pd.DataFrame(data, columns=['Valores'])

# Paso 2: Histograma de Frecuencia Absoluta
plt.figure(figsize=(10, 6))
plt.hist(df['Valores'], bins=np.arange(1, 12) - 0.5, edgecolor='black', alpha=0.7)
plt.title('Histograma de Frecuencia Absoluta')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(np.arange(1, 11))
plt.grid(axis='y')
plt.show()

# Paso 3: Diagrama de Barras Horizontales
frequencies = df['Valores'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
frequencies.plot(kind='barh', color='orange', alpha=0.7)
plt.title('Diagrama de Barras Horizontales')
plt.xlabel('Frecuencia Absoluta')
plt.ylabel('Valores')
plt.grid(axis='x')
plt.show()

# Paso 4: Polígono de Frecuencia Relativa
relative_frequencies = frequencies / frequencies.sum()
plt.figure(figsize=(10, 6))
plt.plot(relative_frequencies.index, relative_frequencies.values, marker='o', linestyle='-')
plt.title('Polígono de Frecuencia Relativa')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Relativa')
plt.xticks(np.arange(1, 11))
plt.grid()
plt.show()

# Paso 5: Diagrama de Pastel de Frecuencia Relativa
plt.figure(figsize=(10, 6))
plt.pie(relative_frequencies, labels=relative_frequencies.index, autopct='%1.1f%%', startangle=90)
plt.title('Diagrama de Pastel de Frecuencia Relativa')
plt.axis('equal')  # Para que el pastel sea circular
plt.show()

# Paso 6: Ojiva (Gráfica de Frecuencia Acumulada)
cumulative_frequencies = frequencies.cumsum()
plt.figure(figsize=(10, 6))
plt.plot(cumulative_frequencies.index, cumulative_frequencies.values, marker='o', linestyle='-')
plt.title('Gráfica de Frecuencia Acumulada (Ojiva)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulada')
plt.xticks(np.arange(1, 11))
plt.grid()
plt.show()
```
